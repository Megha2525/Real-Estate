```python
# Count of missing values per column
missing_values = df.isnull().sum()

# Percentage of missing values per column
missing_percent = (df.isnull().sum() / len(df)) * 100

# Combine into a single DataFrame
missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
}).sort_values(by='Missing Values', ascending=False)

missing_df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=missing_df.index, y=missing_df['Missing Values'], palette='viridis')
plt.xticks(rotation=45)
plt.title('Missing Values per Feature')
plt.ylabel('Number of Missing Values')
plt.xlabel('Features')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=missing_df.index, y=missing_df['Percentage'], palette='magma')
plt.xticks(rotation=45)
plt.title('Percentage of Missing Values per Feature')
plt.ylabel('Percentage (%)')
plt.xlabel('Features')
plt.show()
```

### All histograms or boxplots in one glance

```python
# ===== EDA: Numeric Distributions in a Matrix =====
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Clean numeric columns first (use the safe function from earlier)
numeric_cols = ['price', 'area', 'bedRoom', 'bathroom', 'balcony', 'floorNum', 'agePossession', 'rating']

# Number of rows and columns for the grid
n_cols = 3
n_rows = int(np.ceil(len(numeric_cols)/n_cols))

plt.figure(figsize=(n_cols*5, n_rows*4))

for i, col in enumerate(numeric_cols):
    plt.subplot(n_rows, n_cols, i+1)
    sns.histplot(df[col].dropna(), kde=True, color='skyblue', bins=30)
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('')
    
plt.tight_layout()
plt.show()
```

### Boxplot grid

```python
plt.figure(figsize=(n_cols*5, n_rows*4))

for i, col in enumerate(numeric_cols):
    plt.subplot(n_rows, n_cols, i+1)
    sns.boxplot(x=df[col].dropna(), color='lightgreen')
    plt.title(col)
    plt.xlabel('')
    
plt.tight_layout()
plt.show()
```

### Top categories for each categorical column in a single figure

```python
# ===== EDA: Categorical Features Countplots in a Matrix =====
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

categorical_features = df.select_dtypes(include=['object']).columns

# Set grid size
n_cols = 2  # 2 plots per row for better readability
n_rows = int(np.ceil(len(categorical_features)/n_cols))

plt.figure(figsize=(n_cols*7, n_rows*4))

for i, col in enumerate(categorical_features):
    plt.subplot(n_rows, n_cols, i+1)
    top_categories = df[col].value_counts().nlargest(10).index  # top 10 categories
    sns.countplot(x=df[col], order=top_categories, palette='pastel')
    plt.title(col)
    plt.xticks(rotation=45)
    plt.ylabel('')
    plt.xlabel('')

plt.tight_layout()
plt.show()
```
